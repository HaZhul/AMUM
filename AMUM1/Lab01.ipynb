### Zad 1:

Zaimplementuj funkcję:
$f(x) = \frac{1}{8}x^{4} + 2(x + 0.7)^{3} - \frac{1}{2}x^{2} - 38 x + 1$.



```python
import matplotlib.pyplot as plt
def function(x):
    return (1/8)*pow(x,4) + 2* pow((x + 0.7), 3) - (1/2)*pow(x, 2) -38*x + 1
```

### Zad 2:

Oblicz wartość funkcji $f(x)$ w przedziale [-20, 20] z krokiem $\Delta = 1$. Wyniki zapisz na liście.



```python
y = []
x = range(-20, 21, 1)
for i in x:
    y.append(function(i))
```

### Zad 3:

Wyświetl wykres funkcji $f(x)$ korzystając z wyników z zad. 2.

```python
plt.plot(x, y)
plt.scatter(x, y)
plt.show()
```

### Zad 4:

Korzystając z listy otrzymanej w zad. 2, znajdź takie $x$, dla którego funkcja $f(x)$ przyjmuje najmniejszą wartość. Wyświetl $x$ oraz $f(x)$.

```python
y_min = min(y)
x_min = x[y.index(min(y))]
y_min_index = y.index(y_min) - 20 #ze wzgledu na to, ze x zaczynaja sie od -20

plt.plot(x, y)
plt.scatter(x_min, y_min)
plt.xlim((y_min_index - 5), (y_min_index + 5))
plt.ylim((y_min - 20), (y_min + 200))
plt.show()
```

### Zad 5:

Wyznacz inny krok ($\Delta' < \Delta$) z zakresu [-20, 20] i sprawdź czy odnajdziesz mniejszą wartość funkcji $f(x)$ niż w poprzednim zadaniu? Jeżeli tak, to podaj $x$ oraz wartość funkcji $f(x)$.

```python
x_005 = [x * 0.05 for x in range(-400, 401)]
y_005 = [function(x) for x in x_005]

y_005_min = min(y_005)
x_005_min = x_005[y_005.index(min(y_005))]

print(f"Nowe minimum wynosi: {y_005_min:.4f}")
print(f"Odpowiada ono wartosci x: {x_005_min}")

plt.plot(x_005, y_005)
plt.show()
```

### Zad 6:

Wyświetl wykres (parametry wykresu dobierz samodzielnie) prezentujący minimum funkcji $f(x)$ znalezione w zad. 5.


```python
plt.plot(x_005, y_005)
plt.xlim((y_min_index - 5), (y_min_index + 5))
plt.ylim((y_min - 20), (y_min + 200))
plt.scatter(x_005_min, y_005_min)
plt.show()
```

### Zad 7:

Wygeneruj $k$ losowych wartości $x$ w przedziale [-20, 20](metoda Monte Carlo), a następnie znajdź minimum funkcji $f(x)$.
Wyświetl wykres funkcji.

```python
import random
k = random.randint(400, 800)

x_mc = [random.randint(-400, 401) * 0.05 for x in range(0, k)]
x_mc = sorted(x_mc)
y_mc = [function(x) for x in x_mc]

y_mc_min = min(y_mc)
x_mc_min = x_mc[y_mc.index(min(y_mc))]

print(f"Nowe minimum wynosi: {y_mc_min}")
print(f"Odpowiada ono wartosci x:{x_mc_min}")

plt.plot(x_mc, y_mc)
plt.scatter(x_mc_min, y_mc_min)
plt.show()
```
